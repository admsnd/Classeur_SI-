Quelles étaient les raisons de cet hommage rendu par Google ?

> George Boole a introduit l'algèbre booléenne, un système logique qui est fondamental pour le développement de l'informatique moderne. Le 2 novembre 2015 marquait le 200e anniversaire de sa naissance (né en 1815).

Qu'est-ce qu'une variable booléenne ?

> Une variable booléenne est une variable qui peut prendre seulement deux valeurs : vrai (1) ou faux (0). Ce type de variable est largement utilisé dans la logique et en informatique pour les opérations conditionnelles et de contrôle.

Qu'elles sont les principales fonctions (opérations) booléennes ?

> Les principales opérations booléennes sont :
ET (AND) : La sortie est vraie seulement si les deux entrées sont vraies.
OU (OR) : La sortie est vraie si au moins une des entrées est vraie.
NON (NOT) : Inverse la valeur de l'entrée ; vrai devient faux et vice-versa.
OU exclusif (XOR) : La sortie est vraie si une et une seule des deux entrées est vraie.
