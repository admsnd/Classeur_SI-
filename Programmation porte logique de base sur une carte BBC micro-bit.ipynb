```python
# Code possible pour une fonction ET / AND
from microbit import *

while True:
    if button_a.is_pressed() and button_b.is_pressed():
        display.show(Image.YES)
    else:
        display.show(Image.NO)
```

```python
# Code possible pour une fonction OU / OR
from microbit import *

while True:
    if button_a.is_pressed() or button_b.is_pressed():
        display.show(Image.YES)
    else:
        display.show(Image.NO)
```

```python
# Code possible pour une fonction NON-ET / NAND
from microbit import *

while True:
    if not (button_a.is_pressed() and button_b.is_pressed()):
        display.show(Image.YES)
    else:
        display.show(Image.NO)
```

```python
# Code possible pour une fonction NON-OU / NOR
from microbit import *

while True:
    if not (button_a.is_pressed() or button_b.is_pressed()):
        display.show(Image.YES)
    else:
        display.show(Image.NO)
```

```python
# Code possible pour une fonction OU-EXCLUSIF / XOR
from microbit import *

while True:
    if button_a.is_pressed() != button_b.is_pressed():
        display.show(Image.YES)
    else:
        display.show(Image.NO)
```
